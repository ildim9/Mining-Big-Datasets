## 3) Nearest Neighbor (NN) search

Using the implementation of the previous step, you will calculate the 10-NN (most
similar) customers for the customers with ids listed below (customer id=line number-
1):
`1200, 3650, 10400, 14930, 22330, 25671, 29311, 34650, 39200, 42000`.
For this task your script must take as input the customer-id and `return the list` of her

##### Creation of the lists with the most similar customers for the given customers.
In the code block below we create 10 lists for our given customers.With the first __`for loop`__ we were able to run through them and with the second __`for loop`__ we managed to compute the `dissimilarity` of each one of them with the other 43191 customers.
Because of the mixed data we have in our dataset we had to compute different `dissimilarities` for each type of variable. At the end of the day the 4 different dissimilarities we have computed are the ones below:

   * For the `Nominal` variables we used : $$ d(a,b) = 1~if~a≠b,~0~otherwise $$
    
   * For the `ordinal` variables we used : $$ d(a,b)=\frac{|rank(a)-rank(b)|}{maxRank - minRank} $$
    
   * For the `numeric` variables we used : $$ d(a,b)=\frac{|a-b|}{maxValue - minValue} $$
    
   * For the `set` variables we used :     $$ Jaccard(S1,S2) =1 - \frac{|S1\cap S2|}{|S1\cup S2|} $$  

* The average `Similarity` of the customers is calculated by substructing the 1 from the already computed disimilarities which are stored in the `score` variable.
* At the end of the __`for loop`__ we used the `.append` function in order to fill our lists for each given customer's similarity with the rest of the customers.
* We have also added to our given customers the customer with id = 1 in order to verify our results with the ones given from the assignment's instructions.

* The `Run Time` for the below mentioned procudure is approximately 2 minutes and 30 seconds.

```python
%%time
cust_ids=[1,1200, 3650, 10400, 14930, 22330, 25671, 29311, 34650, 39200, 42000]

similarity_list_1  = []
similarity_list_1200  = []
similarity_list_3650  = []
similarity_list_10400 = []
similarity_list_14930 = []
similarity_list_22330 = []
similarity_list_25671 = []
similarity_list_29311 = []
similarity_list_34650 = []
similarity_list_39200 = []
similarity_list_42000 = []

for c in cust_ids:
   
    for i in range(1,43192):
        age =  abs( banks.Age[c] - banks.Age[i] ) / ( banks.Age.max() - banks.Age.min() )
        job =  [0 if banks.Job[c] == banks.Job[i] else 1]
        marital =   [0 if banks.Marital[c] == banks.Marital[i] else 1]
        education =   abs( banks.Education[c] - banks.Education[i]) / 2 # maxrank - minrank = 3 - 1 = 2 
        default = [0 if banks.Default[c] == banks.Default[i] else 1]
        balance =  abs( banks.Balance[c] - banks.Balance[i] )/ ( banks.Balance.max() - banks.Balance.min() )
        housing =  [0 if banks.Housing[c] == banks.Housing[i] else 1]
        loan =  [0 if banks.Loan[c] == banks.Loan[i] else 1]
        rating =  abs( banks.Rating[c] - banks.Rating[i]) / 4  # maxrank - minrank = 5 - 1 = 4
        
        # Jaccard Dissimilarity
        list1= set(banks.Products[c]) 
        list2= set(banks.Products[i]) 
        
        intersection = len(list1.intersection(list2)) 
        union = len(list1.union(list2)) 
        products = 1 - float(intersection / union)

        score = 1-( age + job + marital + education + default + balance + housing + loan +  rating + products ) / 10
        
        if   c==1200:
            similarity_list_1200.append(score)
        elif c==3650:
            similarity_list_3650.append(score)
        elif c==10400 : 
            similarity_list_10400.append(score)
        elif c==14930 : 
            similarity_list_14930.append(score)
        elif c== 22330 : 
            similarity_list_22330.append(score)
        elif c== 25671 : 
            similarity_list_25671.append(score)
        elif c== 29311 : 
            similarity_list_29311.append(score)
        elif c== 34650 : 
            similarity_list_34650.append(score)
        elif c== 39200 : 
            similarity_list_39200.append(score)
        elif c== 42000 : 
            similarity_list_42000.append(score)
        elif c== 1 : 
            similarity_list_1.append(score)
```

Calculate the similarities between the given customer and all other customers and compute his 10-nn (most similar) customers. __IMPORTANT : In the similarity calculations for this step you need to exclude the customer rating attribute.__

##### Preperatory Steps for the Prediction of the `Rating` .

##### Creation of the lists with the most similar customers for the first 50 customers among all the other customers of the dataset `banks_2`.
In this part of the assignment we implemented the same procedure we did in the 3rd Question. The only diffrence is that we created a list for the every single one of the first 50 customers of our new dataset with the function `customer = [[] for _ in range(n)]`  to save our `similarity` scores for every customer with the others.

At the end of the function we store our similarity scores into the customer list with the use of the `.append()` function. The specific lists contain all the similarities for the first 50 customers with all the others.

The `RunTime` of the procedure below is approximately 12 minutes. 

```python
%%time

customer = [[] for _ in range(51)]

for cust in range(1,51):
       
    for i in range(1,43192):
        
        age = abs( banks_2.Age[cust] - banks_2.Age[i] ) / ( banks_2.Age.max() - banks_2.Age.min() )
        job = [0 if banks_2.Job[cust] == banks_2.Job[i] else 1]
        marital = [0 if banks_2.Marital[cust] == banks_2.Marital[i] else 1]
        education = abs( banks_2.Education[cust] - banks_2.Education[i]) / 2 # maxrank - minrank = 3 - 1 = 2 
        default = [0 if banks_2.Default[cust] == banks_2.Default[i] else 1]
        balance = abs( banks_2.Balance[cust] - banks_2.Balance[i] ) / ( banks_2.Balance.max() - banks_2.Balance.min() )
        housing = [0 if banks_2.Housing[cust] == banks_2.Housing[i] else 1]
        loan =  [0 if banks_2.Loan[cust] == banks_2.Loan[i] else 1]
        
        # Jaccard Dissimilarity
        list1= set(banks_2.Products[cust]) 
        list2= set(banks_2.Products[i]) 

        intersection = len(list1.intersection(list2)) 
        union = len(list1.union(list2)) 
        products = 1 - float(intersection / union)

        score = 1 - ( age + job + marital + education + default + balance + housing + loan  + products ) / 9
        
        customer[cust].append(score)
```

###  4.2) Prediction of the Customer Rating 
Based only on the 10 most similar customers computed in the previous step, predict the customer rating rank using:
* The average rating rank of the 10 most similar customers (rounded to the nearest integer).
* The weighted average rating rank of the 10 most similar customers (rounded to the nearest integer).

$$
Weighted~average~rating~rank = \frac{\sum_{i=1}^{10}[similarity(i)*rank(rating(i))]}{\sum_{i=1}^{10}similarity(i)}
$$

Where:
* `rating(i)` = the rating of the i-th nearest neighbor (i=1 for the most similar customer)
* `similarity(i)` = the similarity of the i-th nearest neighbor with the given customer



##### METHOD 1 : Prediction of customer Rating using the `average rating` of the 10 most similar customers

##### METHOD 2 : Prediction of customer Rating using the __`Weighted Average Rating`__ of the 10 most similar customers

* In order to compute the `weighted average rating` of the 10 most similar customers we have constructed a `for` loop. The loop runs through the datasets which are stored in the `global` variable and computes the weighted average `Rating` of every dataframe ,  rounds it to the nearest integer and stores them into a list called __`average_rating_2`.__
* The `j` runs from 2 to 11 because we exclude the first row which contains the score of the customer himself. It would always be equal with 1.
* The function that computes the weighted average rating has been mentioned in the beginning of the `paragraph 4.2`.

```python
%%time
average_rating_2=[]
for i in range(1,51):
    score_2 = 0
    for j in range(2,12):
        score_2 = score_2 + ( globals()[f"df_{i}"]['Similarity Score'][j] ) * globals()[f"df_{i}"]['Rating'][j] 
    score_2 = (score_2 / sum( globals()[f"df_{i}"]['Similarity Score'])).round()
    average_rating_2.append(score_2)
```

* After the computation of the weighted average rating we tranformed the list into a pandas DataFrame.
